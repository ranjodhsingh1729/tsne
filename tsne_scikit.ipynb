```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

labels_path = "mnist2500_labels.txt"
Labels = labels = np.loadtxt(labels_path)

data_path = "mnist2500_X.txt"
X = pd.read_csv(data_path, header=None, sep="   ", engine="python").to_numpy(dtype=float)
```

```python
tsne = TSNE(
    n_components=2,
    perplexity=30,
    learning_rate=200,
    max_iter=1000,
    n_iter_without_progress=1000,
    random_state=42,
    n_jobs=1,
    verbose=10,
)

Y = tsne.fit_transform(X)
plt.scatter(Y[:, 0], Y[:, 1], 20, labels)
plt.show()
```
