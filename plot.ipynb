```python
import numpy as np
from matplotlib import pyplot as plt

def plot_tsne_results(data_file, labels_file="mnist2500_labels.txt"):
    try:
        Y = np.loadtxt(data_file, delimiter=",")
    except IOError:
        print(f"Error: Could not find the data file '{data_file}'.")
        return

    try:
        labels = np.loadtxt(labels_file)
    except IOError:
        print(f"Error: Could not find the labels file '{labels_file}'.")
        return

    if Y.shape[0] != labels.shape[0]:
        print("Warning: Number of rows in data does not match number of labels. Transposing data.")
        Y = Y.T
    if Y.shape[0] != labels.shape[0]:
        print(f"Error: Mismatch in number of data points ({Y.shape[0]}) and labels ({labels.shape[0]}).")
        return

    print("Got It!")

    plt.scatter(Y[:, 0], Y[:, 1], 20, labels)
    plt.show()
```

```python
data_file = "tsne_result.csv"
plot_tsne_results(data_file)
```
